```python
from scipy.io.wavfile import read
import matplotlib.pyplot as plt
import numpy as np 

# read audio samples
input_data = read("49568_MV.wav")
audio = input_data[1]
# plot the first 1024 samples
audio = audio[4000:12000]/ np.max(audio)
import numpy as np
fs = 4000
a = np.arange(len(audio))
plt.figure(figsize=((13,9)))
plt.plot(a/fs,audio)
# label the axes
plt.ylabel("Amplitude")
plt.xlabel("Time")
# set the title  
plt.title("Sample Wav")
# display the plot
plt.show()
```

```python
from scipy.signal import wiener
```

```python
yy = wiener(audio)

a = np.arange(len(yy))
plt.figure(figsize=((13,9)))
plt.plot(a/fs,audio)
plt.plot(a/fs,yy)
# label the axes
plt.ylabel("Amplitude")
plt.xlabel("Time")
# set the title  
plt.title("Sample Wav")
# display the plot
plt.show()
```
