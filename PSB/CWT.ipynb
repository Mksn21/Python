```python
import plotly.graph_objects as go
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("murmur3.DAT",sep='\s+',header= None)
df=pd.DataFrame(df) 
df = df[:500]
```

```python
x = np.array(df[df.columns[0]])
y = np.array(df[df.columns[1]])

fig = go.Figure(data=go.Scatter(x=x, y=y))
fig.update_layout(
    title="Signal",
    xaxis_title="Time(s)",
    yaxis_title="Amplitudo (mV)"
)
n = len(x)
fig.show()
```

```python
#Libary
from scipy import signal
widths = np.arange(1,101)
fo =0.864
w0 =2*np.pi*fo
cwt = signal.cwt(y, signal.morlet2, widths,w=w0)
cwt= np.absolute(cwt)

fig = go.Figure(data=go.Heatmap(x=x,y=widths,z=cwt))
fig.update_layout(
    title="CWT",
    xaxis_title="time(s)",
    yaxis_title="Scale",
)
fig.show()
```

```python
sc = 0.0001
dsc = 0.0001
fo = 0.849
wo = 2*np.pi *fo

cwtt= np.zeros((100,n))
cwtr = np.zeros((100,n))
cwtim = np.zeros((100,n)) 

for i in range(100): 
    for j in range (n):
        for k in range(n):
            re = 1 / np.sqrt(sc) * 1/ np.power(np.pi,0.25) *np.exp(-(x[k]-x[j])/sc *(x[k]-x[j])/sc/2)*np.sin(wo*(x[k]-x[j])/sc)
            im = 1 / np.sqrt(sc) * 1/ np.power(np.pi,0.25) *np.exp(-(x[k]-x[j])/sc *(x[k]-x[j])/sc/2)*np.cos(wo*(x[k]-x[j])/sc)
            cwtr[i,j] += y[k] *re 
            cwtim[i,j] += y[k] *im 
        cwtt[i,j] = np.sqrt(np.square(cwtr[i,j])+ np.square(cwtim[i,j]))
    sc+=dsc
        
```

```python

fig = go.Figure(data=go.Heatmap(x=x,y=widths,z=cwtt))
fig.update_layout(
    title="CWT",
    xaxis_title="time(s)",
    yaxis_title="Scale",
)
fig.show()
```
